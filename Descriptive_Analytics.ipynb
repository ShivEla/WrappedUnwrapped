Temporal Analysis

Which Day most streamed?

```python
df['weekday'] = df['weekday'].str.replace('\n', '')
```

```python
day=['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
```

```python
min_day=df.groupby('weekday')['minsPlayed'].sum().reindex(day)
```

```python
plt.plot(min_day,color='green',marker="*")
plt.ylabel("Minutes listened")
plt.xlabel("Days of the week")
```

Which day for each artist?

```python
df.groupby('artistName')['weekday'].value_counts()
```

On which day did you listen to Rahman the most?

```python
df.groupby('artistName')['weekday'].value_counts().max()
```

Top 10 songs streamed

```python
df.groupby('trackName')['minsPlayed'].sum().head(10).plot(kind="bar",color='pink')
plt.xticks(rotation=45,ha='right')
plt.title("Top 10 based on duration played")
plt.ylabel("Minutes played")
plt.xlabel('Song Titles')
```

```python
df.trackName.value_counts().head(10)
```

```python
top_10_songs = df.trackName.value_counts().head(10)
colors = plt.cm.rainbow(np.linspace(0, 1, len(top_10_songs)))
df.trackName.value_counts().head(10).plot(kind="bar",color=colors)
plt.xticks(rotation=45,ha='right')
plt.title("Top 10 based on #times played")
plt.ylabel("#times played")
plt.xlabel('Song Titles')
```

Top 20 Streamed Artists

```python
df.artistName.value_counts().head(20).plot(kind='barh')
plt.title("Top 20 based on #times played")
plt.ylabel("#times played")
plt.xlabel('Artists')
```
